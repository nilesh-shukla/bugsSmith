```python
import pandas as pd
import numpy as np
```

```python
USERS_FILE = './dataset-1/users.csv'
FUSERS_FILE = './dataset-1/fusers.csv'
PROCESSED_OUTPUT_FILE = './output-files/processed_users.csv'
```

```python
FOLLOWERS_COL = 'followers_count'
FOLLOWING_COUNT = 'friends_count'
PROFILE_IMG_COL = 'default_profile_image'
PROTECTED_COL = 'protected'
NAME_COL = 'name'
POSTS_COL = 'statuses_count'
```

```python
df_users = pd.read_csv(USERS_FILE, encoding='latin1')
df_users['is_fake'] = 0
df_fusers = pd.read_csv(FUSERS_FILE, encoding='latin1')
df_fusers['is_fake'] = 1
```

```python
df_users.sample(5)
```

```python
df_fusers.sample(5)
```

```python
print(f"Genuine profiles loaded: {len(df_users)}")
print(f"Fake profiles loaded: {len(df_fusers)}")
```

```python
combined_df = pd.concat([df_users, df_fusers], ignore_index=True)
combined_df.sample(10)
```

```python
combined_df[NAME_COL] = combined_df[NAME_COL].fillna('unknown')
combined_df['name_length'] = combined_df[NAME_COL].apply(len)
```

```python
combined_df['follower_ratio'] = combined_df[FOLLOWERS_COL] / (combined_df[FOLLOWING_COUNT]+1)
```

```python
combined_df['protected'] = combined_df[PROTECTED_COL].fillna(0)
combined_df['default_profile_image'] = combined_df[PROFILE_IMG_COL].fillna(0)
```

```python
# Normalize column names to lowercase for robustness

combined_df.columns = [c.lower() for c in combined_df.columns]

if 'id' not in combined_df.columns:
    for alt in ('user_id','uid','profile_id'):
        if alt in combined_df.columns:
            combined_df['id'] = combined_df[alt]
            break
    else:
        combined_df['id'] = range(1, len(combined_df) + 1)

if 'screen_name' not in combined_df.columns:
    for alt in ('handle','username','screenname'):
        if alt in combined_df.columns:
            combined_df['screen_name'] = combined_df[alt]
            break
    else:
        combined_df['screen_name'] = ''

if 'name' not in combined_df.columns:
    combined_df['name'] = combined_df.get('full_name', pd.Series(['unknown']*len(combined_df)))

combined_df['name'] = combined_df['name'].fillna('unknown')

numeric_cols = ['followers_count','friends_count','statuses_count','default_profile_image','protected']
for col in numeric_cols:
    combined_df[col] = pd.to_numeric(combined_df.get(col, 0), errors='coerce').fillna(0).astype(int)

combined_df['name_length'] = combined_df['name'].apply(lambda x: len(str(x)))
combined_df['follower_ratio'] = combined_df['followers_count'] / (combined_df['friends_count'] + 1)

FINAL_FEATURES = [
    'id',
    'name',
    'screen_name',
    'default_profile_image',
    'statuses_count',
    'followers_count',
    'friends_count',
    'protected',
    'name_length',
    'follower_ratio',
    'is_fake',
 ]

final_data = combined_df[FINAL_FEATURES]
final_data.to_csv(PROCESSED_OUTPUT_FILE, index=False)
print(f"\nFinal cleaned dataset for training saved to {PROCESSED_OUTPUT_FILE}.")
final_data.head()
```
